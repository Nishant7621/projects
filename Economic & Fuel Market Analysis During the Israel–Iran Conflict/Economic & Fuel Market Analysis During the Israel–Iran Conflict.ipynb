```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Iran_War_Global_Fuel_Crisis_Dataset.csv")
df.head()
```

```python
df.info()

df.shape

df.describe()
```

```python
df['Date'] = pd.to_datetime(df['Date'])

df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Month_Name'] = df['Date'].dt.month_name()
```

```python
df.dtypes
```

```python
df.duplicated().sum()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['Date'].min(), df['Date'].max()
```

```python
df['Country'].nunique()
```

```python
df['Country'].unique()
```

```python
df['Country'].value_counts()
```

```python
df['Average_Oil_Price'] = (
    df['WTI_Crude_USD_per_barrel']
    + df['Brent_Crude_USD_per_barrel']
    + df['OPEC_Basket_USD_per_barrel']
) / 3

df['Economic_Stress_Index'] = (
    df['Inflation_Rate_Percent']
    + df['Fuel_Price_Change_Percent']
    + df['Freight_Rate_Change_Percent']
) / 3
```

```python
df.head()
```

```python
df.describe().T
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))

sns.heatmap(
    df.select_dtypes(include='number').corr(),
    annot=True,
    cmap='coolwarm'
)

plt.title('Correlation Matrix')
plt.show()
```

```python
top_inf = (
    df.groupby('Country')['Inflation_Rate_Percent']
      .mean()
      .sort_values(ascending=False)
      .head(10)
)

top_inf.plot(kind='bar', figsize=(10,5))
plt.title('Top Countries by Inflation')
plt.show()
```

```python
plt.figure(figsize=(12,5))

daily_fuel = df.groupby('Date')['Fuel_Price_Local'].mean()

plt.plot(daily_fuel)

plt.title('Fuel Price Trend')
plt.xlabel('Date')
plt.ylabel('Fuel Price')
plt.show()
```

```python
plt.figure(figsize=(12,5))

daily_crisis = df.groupby('Date')['Crisis_Intensity_Index'].mean()

plt.plot(daily_crisis)

plt.title('Crisis Intensity Over Time')
plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    x='News_Sentiment_Score',
    y='Stock_Market_Index_Change',
    data=df
)

plt.title('News Sentiment vs Stock Market')
plt.show()
```
